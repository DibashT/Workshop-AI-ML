```python
#Task 2
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Load the grayscale image
image = "/content/cameraman.png"
gray_image = Image.open(image).convert("L")
```

```python
# Display the original grayscale image
plt.figure(figsize=(10, 6))

plt.imshow(gray_image, cmap="gray")
plt.title("Original Grayscale Image")
```

```python
# Convert to NumPy array for processing
gray_array = np.array(gray_image)

# Extract and display the middle 150-pixel section
height, width = gray_array.shape
start_y = height // 2 - 75
cropped_section = gray_array[start_y:start_y + 150, :]

plt.imshow(cropped_section, cmap="gray")
plt.title("Middle 150-Pixel Section")
```

```python
# Apply thresholding (Binary Image)
threshold_value = 100
binary_array = np.where(gray_array < threshold_value, 0, 255).astype(np.uint8)
binary_image = Image.fromarray(binary_array)

plt.imshow(binary_image, cmap="gray")
plt.title("Binary Image (Thresholding)")
```

```python
# Rotate 90 degrees clockwise
rotated_image = gray_image.rotate(-90, expand=True)  # -90 for clockwise rotation


plt.imshow(rotated_image, cmap="gray")
plt.title("Rotated 90 Clockwise")
```

```python
# Convert grayscale to RGB
rgb_image = Image.merge("RGB", (gray_image, gray_image, gray_image))


plt.imshow(rgb_image)
plt.title("Converted to RGB")
```

```python
# Show all images in a grid
plt.tight_layout()
plt.show()
```

```python
#Task 3
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 1. Load and Prepare Data
image_path = "/content/lena_gray.gif"
img = Image.open(image_path).convert("L")
img_array = np.array(img, dtype=np.float64)
```

```python
# Standardize the Data (Centering)
mean = np.mean(img_array, axis=0)
standardized_img = img_array - mean

# Compute Covariance Matrix
cov_matrix = np.cov(standardized_img, rowvar=False)

# 2. Eigen Decomposition
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
```

```python
# Sort eigenvalues & eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Cumulative sum plot to choose k
cumulative_explained_variance = np.cumsum(eigenvalues) / np.sum(eigenvalues)
```

```python
# 3. Compression and Reconstruction with different k values
def compress_and_reconstruct(k):
    """
    Compress and reconstruct the image using the top k principal components.
    """
    top_k_eigenvectors = eigenvectors[:, :k]
    lower_dim_representation = np.dot(standardized_img, top_k_eigenvectors)
    reconstructed_img = np.dot(lower_dim_representation, top_k_eigenvectors.T) + mean
    return np.clip(reconstructed_img, 0, 255).astype(np.uint8)
```

```python
# Experiment with different k values
k_values = [10, 50, 100, 200]

plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.imshow(img_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

for i, k in enumerate(k_values):
    compressed_img = compress_and_reconstruct(k)
    plt.subplot(2, 3, i + 2)
    plt.imshow(compressed_img, cmap="gray")
    plt.title(f"PCA Reconstruction (k={k})")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
